```python
# TIC TAC TOE

# NON-INTERACTIVE CODE CELL. YOU MAY RUN THIS CELL, BUT DO NOT EDIT IT.
# FOR DEMONSTRATION PURPOSES ONLY. DO NOT EDIT.

board1 = [
['X','X','O'],
['O','X','O'],
['O','','X'],
]

board2 = [
['X','X','O'],
['O','X','O'],
['','O','X'],
]

board3 = [
['O','X','O'],
['','O','X'],
['X','X','O'],
]

board4 = [
['X','X','X'],
['O','X','O'],
['O','','O'],
]

board5 = [
['X','X','O'],
['O','X','O'],
['X','','O'],
]

board6 = [
['X','X','O'],
['O','X','O'],
['X','',''],
]

# MAIN FUNCTION

def tic_tac_toe(board):
    # Initialize some variables
    a = 0
    b = 0
    c = len(board) - 1
    d = 0
    i = 0
    j = 0
    board_rows = [x for x in board]
    board_columns = [x for x in zip(*board)]
    board_diagonal_up_down = []
    board_diagonal_down_up = []

    while a < len(board) and b < len(board): # Fill in the diagonal_up_down list
        board_diagonal_up_down.append(board[a][b])
        a = a + 1
        b = b + 1
    
    while c >= 0 and d < len(board): # Fill in the diagonal down up list
        board_diagonal_down_up.append(board[c][d])
        c = c - 1
        d = d + 1
    
    while i < len(board_rows): # Check for horizontal winners
        winner_checker_X = all([letter=="X" for letter in board_rows[i]])
        winner_checker_O = all([letter=="O" for letter in board_rows[i]])
        if winner_checker_X == True:
            return "X"
        elif winner_checker_O == True:
            return "O"
        i = i + 1

    while j < len(board_columns): # Check for vertical winners
        winner_checker_X = all([letter=="X" for letter in board_columns[j]])
        winner_checker_O = all([letter=="O" for letter in board_columns[j]])
        if winner_checker_X == True:
            return "X"
        elif winner_checker_O == True:
            return "O"
        j = j + 1

    # Check for diagonal winners (top to bottom)
    winner_checker_X = all([letter == "X" for letter in board_diagonal_up_down])
    winner_checker_O = all([letter == "O" for letter in board_diagonal_up_down])
    if winner_checker_X == True:
        return "X"
    elif winner_checker_O == True:
        return "O"

    # Check for diagonal winners (bottom to top)
    winner_checker_X = all([letter == "X" for letter in board_diagonal_down_up])
    winner_checker_O = all([letter == "O" for letter in board_diagonal_down_up])
    if winner_checker_X == True:
        return "X"
    elif winner_checker_O == True:
        return "O"

    return None # If there are no winners, return None
```

```python
# ROUTING

# NON-INTERACTIVE CODE CELL. YOU MAY RUN THIS CELL, BUT DO NOT EDIT IT.
# FOR DEMONSTRATION PURPOSES ONLY. DO NOT EDIT.

legs = {
     ("upd","admu"):{
         "travel_time_mins":10
     },
     ("admu","dlsu"):{
         "travel_time_mins":35
     },
     ('dlsu','upd'):{
         "travel_time_mins":55
     }
}

# MAIN FUNCTION

def eta(legs, source, destination):
    previous_location = current_location = source
    time_taken = 0
    for i in legs:
        if i[0] == previous_location:
            current_location = i[1]
            time_taken = time_taken + legs[(previous_location, current_location)]["travel_time_mins"]
            if current_location == destination:
                return time_taken
            previous_location = current_location
eta(legs, "admu", "upd")
```
